```python
# Create an empty list to store unique customer IDs
unique_customer_ids = []

# Function to generate a unique customer ID
def generate_unique_customer_id():
    while True:
        customer_id = random.randint(1, 10000)  # Adjust the range as needed
        if customer_id not in unique_customer_ids:
            unique_customer_ids.append(customer_id)
            return customer_id
```

```python
fake=Faker()


customer_data = pd.DataFrame(columns=["customer_id", "first_name", "last_name", 
                                      "gender", "date_of_birth", "job_title", 
                                      "job_industry_category", "wealth_segment", 
                                      "owns_car"])

# Generate customer data for a specified number of customers
num_customers = 1000  # Adjust as needed

for _ in range(num_customers):
    customer_id = generate_unique_customer_id()
    customer_data = customer_data.append({
        "customer_id": customer_id,
        "first_name": fake.first_name(),
        "last_name": fake.last_name(),
        "gender": random.choice(["Male", "Female"]),
        "date_of_birth": fake.date_of_birth(minimum_age=18, maximum_age=70).strftime("%Y-%m-%d"),
        "job_title": fake.job(),
        "job_industry_category": fake.random_element(elements=("Finance", "Health", "Retail", "IT", "Education")),
        "wealth_segment": fake.random_element(elements=("High Net Worth", "Mass Customer", "Affluent")),
        "owns_car": random.choice(["Yes", "No"])
    }, ignore_index=True)
```

```python
customer_address_data = pd.DataFrame(columns=["customer_id", "address", "postcode", "state", "country"])

# Generate customer address data for the same set of customers
for customer_id in unique_customer_ids:
    customer_address_data = customer_address_data.append({
        "customer_id": customer_id,
        "address": fake.street_address(),
        "postcode": fake.postcode(),
        "state": fake.state_abbr(),
        "country": fake.country()
    }, ignore_index=True)
```

```python
transaction_data = pd.DataFrame(columns=["transaction_id", "customer_id", "transaction_date", "amount", "state"])

# Generate transaction data for a specified number of transactions
num_transactions = 10000  # Adjust as needed

for transaction_id in range(1, num_transactions + 1):
    customer_id = random.choice(unique_customer_ids)
    transaction_date = fake.date_between(start_date="-1y", end_date="today").strftime("%Y-%m-%d")
    amount = round(random.uniform(10, 500), 2)
    state = fake.random_element(elements=("NSW", "VIC", "QLD"))

    transaction_data = transaction_data.append({
        "transaction_id": transaction_id,
        "customer_id": customer_id,
        "transaction_date": transaction_date,
        "amount": amount,
        "state": state
    }, ignore_index=True)
```
